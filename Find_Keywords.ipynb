```python
from rake_nltk import Rake
```

```python
import nltk
```

```python
nltk.download("stopwords")
```

```python
r = Rake()
```

```python
mytext = '''LinkedIn is all about networking and building connections, so it’s important to share valuable content consistently. This could be blog posts, articles, videos, images, or even curated content from other sources.The key is to provide value and insight to your audience, rather than just promoting your products or services.'''
```

```python
nltk.download('punkt')
```

```python
r.extract_keywords_from_text(mytext)
```

```python
r.get_ranked_phrases()
```

```python
r.get_ranked_phrases_with_scores()
```

```python
for ratings,keyword in r.get_ranked_phrases_with_scores():
  if ratings>6:
    print(ratings,keyword)

```

```python
from requests_html import HTMLSession
```

```python
def extract_text():
  s = HTMLSession()
  url = "https://www.semrush.com/blog/linkedin-marketing/"
  response=s.get(url)
  para = response.html.find('p')
  text_content = ' '.join([p.text for p in para])
  return text_content
```

```python
r = Rake()
```

```python
r.extract_keywords_from_text(extract_text())
```

```python
r.get_ranked_phrases()
```

```python
r.get_ranked_phrases_with_scores()
```

```python
for ratings,keywords in r.get_ranked_phrases_with_scores():
  if ratings>8:
    print(ratings,keywords)
```
